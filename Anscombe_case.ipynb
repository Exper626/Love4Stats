Francis Anscombe, a statistician, discovered "Anscombe Quartet". So what's the deal about it?
Here is the deal:
The **Anscombe quartet** has 4 datasets, each set has 11 points.

```python
import pandas as pd

# change in Jupytr:
df = pd.read_csv("anscombe.csv")

# make it a series:
#1:
x1 = df['x1']
y1 = df['y1']

#2:
x2 = df['x2']
y2 = df['y2']

#3:
x3 = df['x3']
y3 = df['y3']

#4:
x4 = df['x4']
y4 = df['y4']

# make the dataframe for each set:
#1:
df1 = pd.DataFrame({"x1": x1, "y1": y1})
print("DataFrame 1 (Dataset I):\n", df1)

#2:
df2 = pd.DataFrame({"x2": x2, "y2": y2})
print("DataFrame 2 (Dataset I):\n", df2)

#3:
df3 = pd.DataFrame({"x3": x3, "y3": y3})
print("DataFrame 3 (Dataset I):\n", df3)

#4:
df4 = pd.DataFrame({"x4": x4, "y4": y4})
print("DataFrame 4 (Dataset I):\n", df4)
```

All four sets have **almost the same average and variance** (which are statistical properties)

```python
# descrriptive summary and shape

#1:
summary_stats = df1.describe()
print(summary_stats)
num_rows, num_columns = df1.shape
print(df1.shape)

#2:
summary_stats = df2.describe()
print(summary_stats)
num_rows, num_columns = df2.shape
print(df2.shape)

#3:
summary_stats = df3.describe()
print(summary_stats)
num_rows, num_columns = df3.shape
print(df3.shape)

#4:
summary_stats = df4.describe()
print(summary_stats)
num_rows, num_columns = df4.shape
print(df4.shape)
```

**BUT!** they look totally different when you plot them.

```python
import matplotlib.pyplot as plt

#1:
plt.scatter(x=df1['x1'], y=df1['y1'], marker='o', label='Dataset I')
plt.xlabel('x1')
plt.ylabel('y1')
plt.title('Scatter Plot: Dataset I')
plt.grid(True)
plt.legend()
plt.show()

#2:
plt.scatter(x=df2['x2'], y=df2['y2'], marker='o', label='Dataset II')
plt.xlabel('x2')
plt.ylabel('y2')
plt.title('Scatter Plot: Dataset II')
plt.grid(True)
plt.legend()
plt.show()

#3:
plt.scatter(x=df3['x3'], y=df3['y3'], marker='o', label='Dataset III')
plt.xlabel('x3')
plt.ylabel('y3')
plt.title('Scatter Plot: Dataset III')
plt.grid(True)
plt.legend()
plt.show()

#4:
plt.scatter(x=df4['x4'], y=df4['y4'], marker='o', label='Dataset IV')
plt.xlabel('x4')
plt.ylabel('y4')
plt.title('Scatter Plot: Dataset IV')
plt.grid(True)
plt.legend()
plt.show()
```

Anscombe wanted to show that graphs matter. When you analyze data, don’t just rely on numbers—**look at the pictures** too!

Basic stats might not capture the real world. So, **visualize** before diving into analysis.
